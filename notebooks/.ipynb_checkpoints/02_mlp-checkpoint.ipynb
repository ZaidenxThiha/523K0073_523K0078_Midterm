## 2. Define the MLP modelReference architecture: Embedding → MeanPool → [Linear → ReLU → Dropout] × L → Linear (output).

```python
def make_model(embed_dim: int, hidden_dims, dropout: float):    model = MLPClassifier(        vocab_size=len(vocab),        embed_dim=embed_dim,        hidden_dims=hidden_dims,        num_classes=2,        dropout=dropout,    )    return model.to(device)# quick sanity checksample_model = make_model(embed_dim=128, hidden_dims=[256, 128], dropout=0.3)sample_model
```

## 3. Training helperWraps training, validation, and final test evaluation for easy sweeps.

```python
def run_experiment(embed_dim: int, hidden_dims, dropout: float, lr: float = 1e-3, weight_decay: float = 1e-5, epochs: int = 5, run_name: str = "exp"):    model = make_model(embed_dim=embed_dim, hidden_dims=hidden_dims, dropout=dropout)    history = train_model(        model,        train_loader=train_loader,        val_loader=val_loader,        epochs=epochs,        lr=lr,        weight_decay=weight_decay,        device=device,    )    test_loss, test_acc = evaluate(model, test_loader, torch.nn.CrossEntropyLoss(), device)    result = {        "run": run_name,        "embed_dim": embed_dim,        "hidden_dims": list(hidden_dims),        "dropout": dropout,        "test_loss": test_loss,        "test_acc": test_acc,        "history": history,    }    return result
```
